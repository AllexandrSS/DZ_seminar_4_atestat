2. Постройте гистограмму для анализа распределения площади жилых

```python
data = pd.read_csv('kc_house_data_4.csv')
```

##### Постройте гистограмму, отображающую распределение цен на недвижимость в наборе данных из файла kc-house-data.

```python
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=30, edgecolor='black')
plt.title('Распределение цен на недвижимость')
plt.xlabel('Цена (USD)')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

##### Постройте гистограмму для анализа распределения площади жилыхпомещений. На графике должно быть ясно видно, какие значения площади наиболее распространены.

```python
plt.figure(figsize=(10, 6))
plt.hist(data['sqft_living'], bins=30, edgecolor='black')
plt.title('Распределение жилой площади')
plt.xlabel('Жилая площадь (sqft)')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

##### Постройте гистограмму, показывающую распределение годов постройки домов. График должен помочь выявить тенденции в строительстве по годам.

```python
plt.figure(figsize=(10, 6))
plt.hist(data['yr_built'], bins=30, edgecolor='black')
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

### Задача 2. Анализ характеристик домов

1. Постройте столбчатую диаграмму для анализа количества домов с видом на
набережную и без вида. График должен показать соотношение между домами с
видом и без него.
2. Постройте столбчатую диаграмму, отображающую распределение домов по
количеству этажей. График должен наглядно показать, как часто встречаются
дома с разным количеством этажей.
3. Постройте столбчатую диаграмму для анализа распределения состояния
домов. График должен показывать частоту встречаемости различных состояний
домов.

##### Постройте столбчатую диаграмму для анализа количества домов с видом нанабережную и без вида. График должен показать соотношение между домами свидом и без него.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='waterfront', hue='waterfront', palette=['skyblue', 'salmon'])
plt.title('Распределение домов по наличию вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Количество домов')
plt.xticks([0, 1], ['Нет', 'Да'], rotation=0)
plt.grid(axis='y')
plt.show()
```

##### Постройте столбчатую диаграмму, отображающую распределение домов по количеству этажей. График должен наглядно показать, как часто встречаются дома с разным количеством этажей.

```python
plt.figure(figsize=(8, 5))
data['floors'].value_counts().sort_index().plot(kind='bar', color='lightgreen')
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.grid(axis='y')
plt.show()
```

##### Постройте столбчатую диаграмму для анализа распределения состояния домов. График должен показывать частоту встречаемости различных состояний домов.


```python
plt.figure(figsize=(8, 5))
data['condition'].value_counts().sort_index().plot(kind='bar', color='coral')
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.grid(axis='y')
plt.show()
```

### Задача 3. Анализ влияния характеристик на стоимость недвижимости

количеством спален и стоимостью недвижимости. График должен показать, как
изменение количества спален влияет на цену.
2. Постройте график размаха (boxplot) для анализа влияния состояния дома на
его стоимость. График должен продемонстрировать, как состояние дома
связано с его ценой и выявить возможные отклонения.
3. Постройте график размаха (boxplot), чтобы исследовать, как количество
этажей влияет на стоимость недвижимости. График должен показать, как
ценовые данные варьируются в зависимости от этажности.
4. Постройте график размаха (boxplot), чтобы проанализировать, как наличие
вида на набережную влияет на стоимость недвижимости. График должен
показать, есть ли значительное влияние наличия вида на цену.
5. Постройте диаграмму рассеяния, чтобы исследовать зависимость между
площадью жилой зоны и стоимостью недвижимости. График должен показать,
как площадь жилой зоны коррелирует с ценой.


##### Постройте диаграмму рассеяния, чтобы изучить зависимость между количеством спален и стоимостью недвижимости. График должен показать, как изменение количества спален влияет на цену.

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['bedrooms'], data['price'], alpha=0.5, color='blue')
plt.title('Цена недвижимости в зависимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена (USD)')
plt.grid(True)
plt.show()
```

##### Постройте график размаха (boxplot) для анализа влияния состояния дома на его стоимость. График должен продемонстрировать, как состояние дома связано с его ценой и выявить возможные отклонения.

```python
plt.figure(figsize=(12, 6))
plt.boxplot([data[data['condition'] == i]['price'] for i in sorted(data['condition'].unique())], 
tick_labels=sorted(data['condition'].unique()))
plt.title('Цена недвижимости в зависимости от состояния')
plt.xlabel('Состояние дома')
plt.ylabel('Цена (USD)')
plt.grid(True)
plt.show()
```

##### Постройте график размаха (boxplot), чтобы исследовать, как количество этажей влияет на стоимость недвижимости. График должен показать, как ценовые данные варьируются в зависимости от этажности.

```python
plt.figure(figsize=(12, 6))
plt.boxplot([data[data['floors'] == i]['price'] for i in sorted(data['floors'].unique())], 
tick_labels=sorted(data['floors'].unique()))
plt.title('Цена недвижимости в зависимости от этажности')
plt.xlabel('Количество этажей')
plt.ylabel('Цена (USD)')
plt.grid(True)
plt.show()
```

##### Постройте график размаха (boxplot), чтобы проанализировать, как наличие вида на набережную влияет на стоимость недвижимости. График должен показать, есть ли значительное влияние наличия вида на цену.

```python
plt.figure(figsize=(12, 6))
data.boxplot(column='price', by='waterfront', grid=False)
plt.title('Цена недвижимости в зависимости от наличия вида нанабережную')
plt.suptitle('') # Удаление дополнительного заголовка,добавляемого pandas)
plt.xlabel('Вид на набережную')
plt.ylabel('Цена (USD)')
plt.xticks([1, 2], ['Нет', 'Да'])
plt.grid(True)
plt.show()
```

##### Постройте диаграмму рассеяния, чтобы исследовать зависимость между площадью жилой зоны и стоимостью недвижимости. График должен показать, как площадь жилой зоны коррелирует с ценой. 

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['sqft_living'], data['price'], alpha=0.5, color='green')
plt.title('Цена недвижимости в зависимости от площади жилой зоны')
plt.xlabel('Жилая площадь (sqft)')
plt.ylabel('Цена (USD)')
plt.grid(True)
plt.show()
```
